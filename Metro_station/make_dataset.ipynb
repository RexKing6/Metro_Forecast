```python
import numpy as np
import pandas as pd
metro_agg = np.zeros((81, 3456, 13))
metro_files = []
for i in range(81):
    metro_files.append(pd.DataFrame(np.zeros((3456, 13))))
for i in range(2, 26):
    metro_file = pd.read_csv('./Metro_train/record_2019-01-'+str(i).zfill(2)+'.csv')
    metro_file.time = pd.to_datetime(metro_file.time)
    start_time = '2019-01-'+str(i)+' 00:00:00'
    start_time = pd.to_datetime(start_time)
    end_time = '2019-01-'+str(i)+' 00:10:00'
    end_time = pd.to_datetime(end_time)
    tenmin_delta = end_time - start_time
    for j in range(144):
        metro_tmp = metro_file[(metro_file.time>=start_time) & (metro_file.time<end_time)].groupby(['stationID', 'status', 'payType']).count()
        for k in range(81):
            metro_files[k].loc[144*(i-2)+j, 0] = str(start_time)
            metro_files[k].loc[144*(i-2)+j, 1] = str(end_time)
            try:
                metro_files[k].loc[144*(i-2)+j, 2] = metro_tmp.loc[k, 1].time.sum()
            except:
                metro_files[k].loc[144*(i-2)+j, 2] = 0
            try:
                metro_files[k].loc[144*(i-2)+j, 3] = metro_tmp.loc[k, 1, 0].time
            except:
                metro_files[k].loc[144*(i-2)+j, 3] = 0
            try:
                metro_files[k].loc[144*(i-2)+j, 4] = metro_tmp.loc[k, 1, 1].time
            except:
                metro_files[k].loc[144*(i-2)+j, 4] = 0
            try:
                metro_files[k].loc[144*(i-2)+j, 5] = metro_tmp.loc[k, 1, 2].time
            except:
                metro_files[k].loc[144*(i-2)+j, 5] = 0
            try:
                metro_files[k].loc[144*(i-2)+j, 6] = metro_tmp.loc[k, 1, 3].time
            except:
                metro_files[k].loc[144*(i-2)+j, 6] = 0
            try:
                metro_files[k].loc[144*(i-2)+j, 7] = metro_tmp.loc[k, 0].time.sum()
            except:
                metro_files[k].loc[144*(i-2)+j, 7] = 0
            try:
                metro_files[k].loc[144*(i-2)+j, 8] = metro_tmp.loc[k, 0, 0].time
            except:
                metro_files[k].loc[144*(i-2)+j, 8] = 0
            try:
                metro_files[k].loc[144*(i-2)+j, 9] = metro_tmp.loc[k, 0, 1].time
            except:
                metro_files[k].loc[144*(i-2)+j, 9] = 0
            try:
                metro_files[k].loc[144*(i-2)+j, 10] = metro_tmp.loc[k, 0, 2].time
            except:
                metro_files[k].loc[144*(i-2)+j, 10] = 0
            try:
                metro_files[k].loc[144*(i-2)+j, 11] = metro_tmp.loc[k, 0, 3].time
            except:
                metro_files[k].loc[144*(i-2)+j, 11] = 0
            if start_time.weekday()<=4:
                metro_files[k].loc[144*(i-2)+j, 12] = 0
            else:
                metro_files[k].loc[144*(i-2)+j, 12] = 1
        start_time = end_time
        end_time += tenmin_delta
```

```python
for i in range(81):
    metro_files[i].rename(columns={0:'start_time', 1:'end_time', 2:'in_num', 3:'in_paytype_0', 4:'in_paytype_1', 5:'in_paytype_2', 6:'in_paytype_3', 
                      7:'out_num', 8:'out_paytype_0', 9:'out_paytype_1', 10:'out_paytype_2', 11:'out_paytype_3', 12:'weekend'}, inplace=True)
    metro_files[i].to_csv('station-'+str(i).zfill(2)+'.csv', index=False)
```
